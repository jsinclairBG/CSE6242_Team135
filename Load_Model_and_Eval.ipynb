```python
def load_image(path):
  img = Image.open(path)

  transform = T.Compose([
            T.ToTensor(),
            T.Resize((64, 64)),
            T.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225)),
        ])
  
  tensor_image = transform(img).unsqueeze(0)
  return tensor_image
```

```python
x = load_image('test_image.jpg')
```

```python
model = torch.load('pawpularity_sigmoid_model', map_location=torch.device('cpu'))

model.eval()

with torch.no_grad():
  output = model(x) * 100

output
```

```python
face = torch.load('face_detection', map_location=torch.device('cpu'))

face.eval()

with torch.no_grad():
  face_score = face(x)

face_score
```
